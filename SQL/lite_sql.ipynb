```python
%%sql

SELECT tracks.Name, tracks.Composer, media_types.Name as MediaTypeName
FROM tracks
JOIN media_types
    ON tracks.MediaTypeId = media_types.MediaTypeId
LIMIT 10
```

```python
%%sql

SELECT t1.Name, t1.Composer, t1.Milliseconds, t1.Bytes, t1.UnitPrice, t2.Name as MediaType, t3.Title as AlbumTitle, t4.Name as Genre, t5.name as ArtistName
FROM tracks as t1
JOIN media_types as t2
    ON t1.MediaTypeId = t2.MediaTypeId
JOIN albums as t3
    ON t1.AlbumId = t3.AlbumId
JOIN genres as t4
    ON t1.GenreId = t4.GenreId
JOIN artists as t5
    ON t3.ArtistId = t5.ArtistId    
LIMIT 10
```

```python
%%sql

SELECT SUM((t1.Milliseconds)/1000/60) || ' mins ' as DurationMins, 
        ROUND(AVG(t1.Bytes)*0.000001,1) as AverageMegabytes, 
        t1.UnitPrice,
        t5.name as ArtistName
FROM tracks as t1 -- my comment --
JOIN media_types as t2
    ON t1.MediaTypeId = t2.MediaTypeId
JOIN albums as t3
    ON t1.AlbumId = t3.AlbumId
JOIN genres as t4
    ON t1.GenreId = t4.GenreId
JOIN artists as t5
    ON t3.ArtistId = t5.ArtistId
GROUP BY t3.ArtistId   
LIMIT 10
```

```python
%%sql

SELECT *
FROM tracks
WHERE tracks.AlbumId IN (
    SELECT AlbumId
    FROM albums as t1
    JOIN artists as t2
    ON t2.ArtistId = t1.ArtistId
    WHERE Title LIKE "B%"
)
LIMIT 10
```

```python
%%sql


WITH myTemporaryResultName as 
(
    SELECT AlbumId
    FROM albums as t1
    JOIN artists as t2
    ON t2.ArtistId = t1.ArtistId
    WHERE Title LIKE "B%"
)
SELECT *
FROM tracks
WHERE tracks.AlbumId IN myTemporaryResultName
LIMIT 10
```
