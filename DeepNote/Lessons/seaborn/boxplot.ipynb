```
import seaborn as sns
import matplotlib.pyplot as plt
#sns.set(style="whitegrid")
#%matplotlib inline

df = sns.load_dataset("titanic")
df.head(2)
```

# BOXPLOT

A `box plot` (`whisker plot`) shows the distribution of NUMERICAL data based on CATEGORICAL categories

`box` shows the quartiles of the dataset 

`whiskers` extend to show the rest of the distribution, except for points that are determined to be `outliers`

```
sns.boxplot(data=df,
            x="survived", 
            y="age")
```

`linewidth` to make lines wider

```
sns.boxplot(data=df,
            x="survived", 
            y="age",
                linewidth=3,
                order=[1,0])
```

If you want to display all the data points

```
sns.boxplot(data=df,
            x="pclass", 
            y="age")

sns.swarmplot(data=df,
            x="pclass", 
            y="age", color=".25")
```

`orient="h"` for different orientation - might be better to present data this way. You need to switch `x,y` axis appropriately

```
sns.boxplot(data=df,
            x="age", 
            y="survived",
                orient="h")
```

```
plt.figure(figsize=(12,8))
sns.boxplot(data=df,
            x="survived", 
            y="age",
            hue="pclass")
```

# VIOLIN PLOT

everything above can be used with violin plot, moreover it can do following

```
plt.figure(figsize=(12,8))
sns.violinplot(data=df,
                x="pclass", 
                y="age", 
                hue="survived")
```

```
plt.figure(figsize=(12,8))
sns.violinplot(data=df,
                x="pclass", 
                y="age", 
                hue="survived",
                     split=True)
```

It can also display statistical quartiles.

```
plt.figure(figsize=(12,8))
sns.violinplot(data=df,
                x="pclass", 
                y="age", 
                hue="survived", 
                    split=True,
                    inner="quartile")
```
