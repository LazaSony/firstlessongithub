```
#Load Packages
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import urllib.request
import requests
print("Packages loaded")
import time
import random
```

```
max_pages = 20
list_of_image_urls = []

for i in range(1, max_pages+1):
    # get the page
    page_url = "https://www.najsport.sk/bicykle?page=" + str(i)
    request = requests.get(page_url)
    soup = BeautifulSoup(request.content,"lxml")

    # get all bikes on the page
    bike_div = soup.find("div", attrs={"id":"fs_results_box"})
    bikes = bike_div.find_all("img", attrs={"class":"product-card__img"})

    # get all img links from the bikes
    for bike in bikes:
        bike_image_url = "https://www.najsport.sk" + bike["data-src"]
        list_of_image_urls.append(bike_image_url)
    
    # waiting time
    waiting_time = random.uniform(1, 3)
    time.sleep(waiting_time)

# download images
# idx = 1
# for url in list_of_image_urls:
#     with open("/work/data/scraped_images/"+f"image_{idx}.jpg", "wb") as f:
#         f.write(requests.get(url).content)
#         idx += 1
```
